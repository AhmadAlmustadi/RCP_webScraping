```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV file
df = pd.read_csv('/Armin Storm_output.csv')

# Print the DataFrame
df.head(3)
```

```python
# Descriptive statistics
df.describe()
```

```python
df.info()
```

```python
# Check for missing values
missing_values = df.isnull().sum()

# Print the number of missing values in each column
print(missing_values)
```

```python
 #missing values by colums

missing_values_by_column = df.isnull().sum()
print(missing_values_by_column)
```

```python
# Counts of watches by brand
brand_counts = df['brand'].value_counts()
print("Counts of Watches by Brand:")
print(brand_counts)

# Counts of watches by Parent Module
parent_module_counts = df['Parent_Module'].value_counts()
print("Counts of Watches by Parent Module:")
print(parent_module_counts)
```

```python
# Find the most expensive watch
most_expensive_watch = df.loc[df['Price'].idxmax()]

# Find the cheapest watch
cheapest_watch = df.loc[df['Price'].idxmin()]

print("Most Expensive Watch:")
print(most_expensive_watch)

print("\nCheapest Watch:")
print(cheapest_watch)
```

**Data Visualization:**

```python
# Histogram of prices
plt.figure(figsize=(10, 6))
sns.histplot(df['Price'], bins=20, kde=True)
plt.title('Distribution of Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

# Boxplot of prices by brand
plt.figure(figsize=(12, 8))
sns.boxplot(x='Parent_Model', y='Price', data=df)
plt.title('Prices Distribution by Parent Module')
plt.xlabel('Parent Module')
plt.ylabel('Price')
plt.xticks(rotation=45, ha='right')
plt.show()



```
